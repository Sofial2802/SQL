1. List the name and vendor of those products that have not been included in any order.

```python
q="""
SELECT productCode, productName, productVendor
FROM products
WHERE productCode NOT IN (SELECT productCode FROM orderdetails)
"""
pd.read_sql(q,con)
```

2. Create a field that categorizes the quantity of units ordered per order: less than 100, between 100 and 200, or more than 200. If an order has no order lines, it will be labeled as 'No order lines' in this field. How many orders are there of each type?

```python
q="""
WITH aux AS (
  SELECT SUM(od.quantityOrdered) AS tot_units, o.orderNumber,
  CASE WHEN SUM(od.quantityOrdered)<100 THEN "Less than 100"
  WHEN SUM(od.quantityOrdered) BETWEEN 100 AND 200 THEN "Between 100 and 200"
  WHEN SUM(od.quantityOrdered)>200 THEN "More than 200"
  ELSE "No order lines"
  END AS category
  FROM orders o
  LEFT JOIN orderdetails od ON o.orderNumber=od.OrderNumber
  GROUP BY o.orderNumber
)

SELECT COUNT(orderNumber)AS order_count, category
FROM aux
GROUP BY category"""
pd.read_sql(q,con)
```

```python
q = """
WITH aux AS (
SELECT o.orderNumber, SUM(od.quantityOrdered) AS total_units,
CASE WHEN SUM(od.quantityOrdered) < 100 THEN 'Less than 100'
WHEN SUM(od.quantityOrdered) BETWEEN 100 AND 200 THEN 'Between 100 and 200'
WHEN SUM(od.quantityOrdered) > 200 THEN 'More than 200'
ELSE 'No order lines'
END AS category
  FROM orders o
  LEFT JOIN orderdetails od ON o.orderNumber = od.orderNumber
  GROUP BY o.orderNumber
)

SELECT COUNT(orderNumber) AS order_count, category
FROM aux
GROUP BY  category;
"""

pd.read_sql(q, con)
```

3. Create a list of all postal codes where packages can be received or sent. In the case of an office, the 'Action' field will contain 'E', and in the case of a customer's address, it will contain 'R'. The list must not contain duplicates.




```python
q = """
SELECT postalCode, 'R' AS Action
FROM customers
UNION
SELECT postalCode, 'E' AS Action
FROM offices
"""

pd.read_sql(q, con)
```

4. List all the products purchased by Herkku Gifts.

```python
q="""
SELECT p.productCode, p.productName
FROM products p
JOIN orderdetails od ON p.productCode=od.productCode
JOIN orders o ON od.orderNumber=o.orderNumber
JOIN customers c ON o.customerNumber=c.customerNumber
WHERE customerName="Herkku Gifts"
"""
pd.read_sql(q,con)
```

5. Who are the employees in Boston?

```python
q="""
SELECT e.employeeNumber, e.officeCode, e.firstName, e.lastname
FROM employees e
INNER JOIN offices of ON e.officeCode=of.officeCode
WHERE of.city="Boston"
"""
pd.read_sql(q,con)
```

```python
q = """
SELECT employeeNumber, firstName, lastName
FROM employees
WHERE officeCode IN (SELECT officeCode FROM offices WHERE city = 'Boston')
"""

pd.read_sql(q, con)
```

6. Compute the revenue generated by each sales representative based on the orders from the customers they serve.

```python
q="""
SELECT e.employeeNumber,e.firstName,e.lastName, SUM(od.quantityOrdered*od.priceEach)ASrev_generated
FROM employees e
JOIN customers c ON e.employeeNumber=c.salesRepEmployeeNumber
JOIN orders o ON c.customerNumber=o.customerNumber
JOIN orderdetails od ON orderNumber=o.orderNumber
GROUP BY e.employeeNumber,e.firstNumber, e.lastName"""
pd.read_sql(q,con)
```

```python
q = """
SELECT e.employeeNumber, e.firstName, e.lastName, SUM(od.quantityOrdered * od.priceEach) AS revenue
FROM employees e
JOIN customers c ON e.employeeNumber = c.salesRepEmployeeNumber
JOIN orders o ON c.customerNumber = o.customerNumber
JOIN orderdetails od ON o.orderNumber = od.orderNumber
GROUP BY e.employeeNumber, e.firstName, e.lastName
"""

pd.read_sql(q, con)
```

7. Find the products sold in 2003 but not in 2004.

```python
q="""
WITH first AS(
    SELECT p.productName, o.orderDate, p.productCode
    FROM products p
    LEFT JOIN orderdetails od ON od.productCode=p.productCode
    LEFT JOIN orders o ON o.orderNumber=od.orderNumber
    WHERE strftime("%Y",o.orderDate)="2003"
    GROUP BY p.productName),

    second AS(
      SELECT p.productName, o.orderDate, p.productCode
    FROM products p
    LEFT JOIN orderdetails od ON od.productCode=p.productCode
    LEFT JOIN orders o ON o.orderNumber=od.orderNumber
    WHERE strftime("%Y",o.orderDate)="2004"
    GROUP BY p.productName
    )
SELECT *
FROM first
LEFT JOIN second ON first.productName=second.productName
WHERE second.productName IS NULL"""
pd.read_sql(q,con)
```

```python
q = """
WITH i AS (
SELECT p.productName, strftime('%Y', o.orderDate) year
FROM products p
LEFT JOIN orderDetails od ON od.productCode = p.productCode
LEFT JOIN orders o ON o.orderNumber = od.orderNumber
WHERE year = '2003'
GROUP BY p.productName
),

n AS (
SELECT p.productName, strftime('%Y', o.orderDate) year
FROM products p
LEFT JOIN orderDetails od ON od.productCode = p.productCode
LEFT JOIN orders o ON o.orderNumber = od.orderNumber
WHERE year = '2004'
GROUP BY p.productName
)

SELECT *
FROM i
LEFT JOIN n ON i.productName = n.productName
WHERE n.productName IS NULL
"""

pd.read_sql(q, con)
```
