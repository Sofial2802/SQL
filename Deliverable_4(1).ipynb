1. List those customers whose first order had a value of more than â‚¬30,000.

```python
q="""
SELECT customerName
FROM customers
WHERE customerNumber IN(
  SELECT customerNumber
  FROM orders
  WHERE orderNumber IN(
    SELECT MIN(orderNumber)
    FROM orders
    GROUP BY customerNumber
  )
AND orderNumber IN(
  SELECT orderNumber
  FROM orderdetails
  GROUP BY orderNumber
  HAVING SUM(priceEach*quantityOrdered)>30000
)
)
"""
pd.read_sql(q,con)
```

2. List the first name and last name of each manager followed by the first name and last name of each employee they supervise. Sort the results by the last name of each manager.



```python
q="""
WITH manager AS(
  SELECT employeeNumber, lastName,firstName, jobTitle
  FROM employees
  WHERE jobTitle LIKE ("%MANAGER%")
)
SELECT e.employeeNumber, e.lastName, e.firstName, e.reportsTo, m.firstName, m.lastName
FROM employees e
INNER JOIN manager m ON e.reportsTo=m.employeeNumber
ORDER BY m.lastname DESC
"""
pd.read_sql(q,con)
```

3. List the names of the top 10 customers with the greatest number orders, considering only those who have spent more than â‚¬10,000 in the total of their orders.

```python
q = """
WITH filteredCustomers AS(SELECT SUM(od.priceEach*od.quantityOrdered) AS TotalMoney, customerNumber
FROM orderdetails od
INNER JOIN orders o ON od.orderNumber=o.orderNumber
GROUP BY customerNumber
HAVING TotalMoney>10000)

SELECT c.customerName, fc.TotalMoney
FROM filteredCustomers fc
INNER JOIN customers c ON fc.customerNumber=c.customerNumber
ORDER BY fc.TotalMoney DESC LIMIT 10;
"""

pd.read_sql(q,con)
```

4. List the top 3 countries with the most units of product sold.

```python
q="""
WITH countrys AS(
  SELECT c.country, SUM(quantityordered) unitssold
  FROM orderdetails od
  INNER JOIN orders o ON od.orderNumber=o.orderNumber
  INNER JOIN customers c ON o.customerNumber=c.customerNumber
  GROUP BY c.country)
SELECT country, unitssold
FROM countrys
ORDER BY unitssold DESC
LIMIT 3"""
pd.read_sql(q,con)
```

```python
q = """
WITH CountrySales AS (
    SELECT
        c.country,
        SUM(od.quantityOrdered) AS totalUnitsSold
    FROM orderdetails od
    INNER JOIN orders o ON od.orderNumber = o.orderNumber
    INNER JOIN customers c ON o.customerNumber = c.customerNumber
    GROUP BY c.country
)
SELECT country, totalUnitsSold
FROM CountrySales
ORDER BY totalUnitsSold DESC
LIMIT 3
"""

# Execute the query and fetch the result
pd.read_sql(q, con)
```

5. How many customers have placed more than one order on the same day?

```python
q="""
SELECT o.customerNumber, COUNT(DISTINCT o.orderNumber) AS date, orderDate
FROM orders o
LEFT JOIN customers c ON o.customerNumber=c.customerNumber
GROUP BY c.customerNumber, o.orderDate
HAVING date>1"""
pd.read_sql(q,con)
```

```python
q="""
SELECT customerNumber, orderDate, COUNT(DISTINCT orderNUmber)
FROM orders
GROUP BY customerNumber, orderDate
HAVING COUNT(DISTINCT orderNumber)>1"""
pd.read_sql(q,con)
```

6. Create a histogram of orders placed by each customer in 2005.

```python
q="""
SELECT numOrders, COUNT(customerNumber) AS numberOfCustomers
FROM (
SELECT customerNumber, COUNT(orderNumber) AS numOrders
FROM orders
WHERE strftime('%Y', orderDate) = '2005'
GROUP BY customerNumber
) AS orderCounts
GROUP BY numOrders
ORDER BY numOrders;
"""
pd.read_sql(q,con)
```

7. Compute the ratio of payments for each customer vs. the total of payments.

```python
q="""
WITH customerpayment AS(
  SELECT c.customerName, c.customerNumber, SUM(p.amount) customer_payment
  FROM customers c
  LEFT JOIN payments p ON c.customerNumber=p.customerNumber
  GROUP BY c.customerName
),
totalpayments AS(
  SELECT SUM(p.amounts) as total_payments
  FROM payments )
SELECT cp.customerNumber, cp.customer_payment/tp.total_payments AS paymentratio
FROM customerssss cp
CROSS JOIN totalpayment tp"""
pd.read_sql(q,con)
```

```python
q="""
WITH customer_payments AS(
  SELECT c.customerName, SUM(p.amount) AS customerpayment, c.customerNumber
  FROM customers c
  LEFT JOIN payments p ON c.customerNumber=p.customerNumber
  GROUP BY c.customerNumber
), total_payment AS(
  SELECT SUM(p.amount) AS total_payment
  FROM payments p
)
SELECT cp.customerNumber,cp.customerpayment/tp.total_payment AS paymentratio
FROM customer_payments cp
CROSS JOIN total_payment tp """
pd.read_sql(q,con)
```

```python
q="""
WITH total_payments AS (
    SELECT SUM(p.amount) AS total_payment
    FROM payments p
),
customer_payments AS (
    SELECT c.customerNumber, SUM(p.amount) AS customer_payment
    FROM customers c
    LEFT JOIN payments p ON p.customerNumber = c.customerNumber
    GROUP BY c.customerNumber
)
SELECT cp.customerNumber,
       cp.customer_payment / tp.total_payment AS payment_ratio
FROM customer_payments cp
CROSS JOIN total_payments tp"""

pd.read_sql(q,con)
```
