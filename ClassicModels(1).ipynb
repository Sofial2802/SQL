
## Date functions
Extracting Date Components:

  * strftime('%Y', ExampleDate): Returns the year.
  * strftime('%m', ExampleDate): Returns the month.
  * strftime('%d', ExampleDate): Returns the day.
  * strftime('%H', ExampleDate): Returns the hour.
  * strftime('%M', ExampleDate): Returns the minute.
  * strftime('%S', ExampleDate): Returns the second.


For example, with the following query, it would return the order date in the 'mm-YYYY' format.


  ```
  SELECT strftime('%m-%Y', orderDate)
  FROM orders
  ```


Date and Time Arithmetic:

### CASE WHEN AND UNION

The **CASE WHEN** statement is a conditional expression that allows you to perform **conditional logic** within a query. It evaluates a series of conditions and **returns a result based on the first condition that is true**.

* Classify the list of customers based on the payment they have made: create 3 categories, less than 20,000, between 20,000 and 50,000, and more than 50,000. How many customers are there of each type?

````python
q="""
WITH aux AS(
SELECT c.customerNumber,SUM(p.amount),
CASE WHEN SUM(p.amount)<20000 THEN "Less than 2000"
WHEN SUM(p.amount)>=20000 AND SUM(p.amount)<=50000 THEN "Btw 20k and 50k"
WHEN SUM(p.amount)>50000 THEN "More than 50k"
ELSE "No payment"
END AS class
FROM customers c
LEFT JOIN payments p ON p.customerNumber=c.customerNumber
GROUP BY c.customerNumber)

SELECT COUNT(customerNumber), class
FROM aux
GROUP BY class


"""
pd.read_sql(q,con)
````

* How many customers have placed an order and how many have not?

````python
q="""
WITH aux AS
(SELECT DISTINCT c.customerNumber, CASE WHEN o.customerNumber IS NULL THEN "without" ELSE "with" END AS type
FROM customers c
LEFT JOIN orders o ON c.customerNumber=o.customerNumber)
SELECT COUNT(customerNumber), type
FROM aux
GROUP BY type
"""

pd.read_sql(q,con)
````

* Calculate the ratio of shipped orders to the total number of orders for each customer in 2003

## WINDOW FUNCTIONS:
Window functions allow us to perform aggregated operations on a set of rows while maintaining the separate identity of each row in the result.

The structure of the window function is as follows:

* Aggregation or ranking function:
  * SUM(), MAX(), MIN(), AVG(). COUNT()
  * RANK(), DENSE_RANK(), ROW_NUMBER()
* OVER
* PARTITION BY: Field or a list of fields for partitioning (equivalent to the fields specified in the GROUP BY clause)
* ORDER BY: Sorts the rows within each partition
When you use group by the rows are the same, you loose info so. You are loosing the individuality of each one of the rows. If you want to compute the number of orders done by customers.

Able with one operation to compute a number without loosing the value of the others


### Aggregation functions

#### We want to know the number of orders made per city without losing the complete information of each order. The first way to do this exercise would be by using a subquery:

````python
q="""
SELECT
  customerNumber,orderNumber,orderDate,
  COUNT(OrderNumber) OVER( PARTITION BY customerNumber) AS total_orders
FROM
  orders"""
pd.read_sql(q,con)
````

#### This example can be done in a single query using a window function:

````python
q="""
SELECT
  c.customerNumber,orderNumber,orderDate, city,
  COUNT(OrderNumber) OVER( PARTITION BY c.city) AS total_orders
FROM
  orders o
INNER JOIN customers c ON c.customerNumber=o.customerNumber
ORDER BY city"""
pd.read_sql(q,con)
````

### Ranking functions

The three functions will assign a number to our row, but there are differences between them.

* ROW_NUMBER():It will assign continuous numbers, and no number will be repeated in different rows within the same partition.
* RANK():It will assign the same number in case there are rows with the same sorting and partition values. If this happens, it will create gaps within rows of the same partition.
* DENSE_RANK():It will assign the same number in case there are rows with the same sorting and partition values, but it will not have gaps within the partition."

#### Create a ranking of customers with the most orders for each country.

RANK: If two customers from the same country have the same number of orders, they will have the same rank number, and the next customer will consider both rows with the same ranking.

````python
q="""
SELECT COUNT(orderNumber), c.customerNumber, c.country, RANK() OVER (PARTITION BY country ORDER BY COUNT(orderNumber) DESC) AS rnk
FROM orders o
INNER JOIN customers c ON c.customerNumber=o.customerNumber
GROUP BY c.customerNumber

"""
pd.read_sql(q,con)
````

DENSE_RANK: If two customers from the same country have the same number of orders, they will have the same rank number, and the next customer will continue with the sequence.

````python
q="""
SELECT COUNT(orderNumber), c.customerNumber, c.country, RANK() OVER (PARTITION BY country ORDER BY COUNT(orderNumber) DESC) AS rnk
FROM orders o
INNER JOIN customers c ON c.customerNumber=o.customerNumber
GROUP BY c.customerNumber

"""
pd.read_sql(q,con)
````
