1. Number of unique customers

```python
q="""
SELECT DISTINCT customerName
FROM customers
"""
pd.read_sql(q,con)
```

2. List the customers who have NOT placed any orders and who are NOT from the USA.

```python
q="""
SELECT DISTINCT (customerNumber), country
FROM customers
WHERE customerNumber NOT IN (SELECT customerNumber FROM orders) AND country != "USA"

"""
pd.read_sql(q,con)
```

```python
q="""
SELECT DISTINCT customerNumber, customerName, country
FROM customers
WHERE customerNumber NOT IN (SELECT customerNumber FROM orders) AND country!="USA"
"""
pd.read_sql(q,con)
```

3. Number of customers per country for countries with more than 5 customers, from highest to lowest count.




```python
q="""
SELECT country, COUNT(customerNumber)
FROM customers
GROUPBY country
HAVING COUNT(customerNumber)>5
ORDERBY COUNT(customerNumber) DESC"""
pd.read_sql(q,con)
```

```python
q="""
SELECT DISTINCT COUNT(customerNumber), country
FROM customers
GROUP BY country
HAVING COUNT(customerNumber)>5
ORDER BY COUNT(customerNumber) DESC

"""
pd.read_sql(q,con)
```

4. Number of orders recorded in the year 2005 per month and order status

```python
q="""
SELECT strftime("%m",orderDate), status, orderNumber
FROM orders
WHERE strftime("%Y", orderDate)="2005"
GROUP BY status, strftime("%m",orderDate)
ORDER BY strftime("%m", orderDate)
"""
pd.read_sql(q,con)
```

5. List orders that have more than 3 orderdetails related with.

```python
q="""
SELECT orderNumber, COUNT(productCode)
FROM orderdetails
GROUP BY orderNumber
HAVING COUNT(productCode)>3

"""
pd.read_sql(q,con)
```

6. Which employees reports to Mary Patterson?

```python
q="""
SELECT lastName, firstName, reportsTo
FROM employees
WHERE reportsTo="1002"
"""
pd.read_sql(q,con)
```

7. Which payments are more than twice the average payment? Order the results by the date of the payment.

```python
q="""
SELECT amount
FROM(SELECT amount
FROM payments
WHERE amount>(SELECT 2*AVG(amount) FROM payments))

"""
pd.read_sql(q,con)
```
